```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Global_Df = pd.read_csv('Global_Superstore2.csv',encoding='Latin-1')
```

```python
Global_Df.head()
```

```python
Global_Df.dropna(inplace=True)
Global_Df.drop(columns=['Row ID','Order ID','Customer Name','Country','Market'],inplace=True)
Global_Df['Postal Code'] =Global_Df['Postal Code'].astype(str)
```

```python
Global_Df.reset_index(level=False,drop='index',inplace=True)
```

<div style="font-size:30px;text-align:center">Categorical Variables</div>

```python


# Assuming you have a pandas DataFrame called df with categorical variables
categorical_vars = ['Ship Mode', 'Segment', 'Region', 'Category', 'Sub-Category', 'Order Priority']

for var in categorical_vars:
    category_counts = Global_Df[var].value_counts()

    # Create a bar chart or pie chart based on the nature of the variable
    if len(category_counts) <= 10:
        # Bar chart for nominal variables with fewer categories
        category_counts.plot(kind='bar', figsize=(8, 6))
        plt.xlabel(var)
        plt.ylabel('Frequency')
        plt.title(f'{var} Distribution')
        plt.xticks(rotation=0)
        plt.show()
    else:
        # Pie chart for ordinal variables with more categories
        category_counts.plot(kind='pie', figsize=(30, 10), autopct='%0.1f%%',textprops={'fontsize': 12})
        
        plt.ylabel('')
        plt.title(f'{var} Distribution')
        plt.show()
```

```python
# Sales of state
Sales_State  = Global_Df.groupby('State')['Sales'].sum().reset_index()
plt.figure(figsize=(30, 20))

x = Sales_State['State']
y = Sales_State ['Sales']
plt.bar(x,y,color='green')
 # Red color for negative values
plt.xlabel('State',fontsize=25)
plt.ylabel('Sales',fontsize=25)
plt.title('Sales by State',fontsize=30)
plt.xticks(rotation=90,fontsize=20)
plt.yticks(fontsize=20)

plt.show()
```

<div style="font-size:30px; text-align:center">Time period</div>

```python
Global_Df['Ship Date'] = pd.to_datetime(Global_Df['Ship Date'])
Global_Df['Order Date'] = pd.to_datetime(Global_Df['Order Date'])

Global_Df.sort_values(by='Order Date', inplace=True)

plt.plot(Global_Df['Order Date'], Global_Df['Ship Date'])
plt.xlabel('Order Date')
plt.ylabel('Ship Date')
plt.title('Ship Date vs Order Date')
plt.xticks(rotation=45)
plt.show()
```

```python
data_agg = Global_Df.groupby(pd.Grouper(key='Order Date', freq='M')).count()

plt.plot(data_agg.index, data_agg['Ship Date'])
plt.xlabel('Order Date')
plt.ylabel('Number of Shipments')
plt.title('Trends of Shipments over Month')
plt.xticks(rotation=45)
plt.show()
```

```python
data_agg = Global_Df.groupby(pd.Grouper(key='Order Date', freq='Q')).count()

plt.plot(data_agg.index, data_agg['Ship Date'])
plt.xlabel('Order Date')
plt.ylabel('Number of Shipments')
plt.title('Trends of Shipments over Quarters')
plt.xticks(rotation=45)
plt.show()
```

```python
data_agg = Global_Df.groupby(pd.Grouper(key='Order Date', freq='Y')).count()

plt.plot(data_agg.index, data_agg['Ship Date'])
plt.xlabel('Order Date')
plt.ylabel('Number of Shipments')
plt.title('Trends of Shipments over Year')
plt.xticks(rotation=45)
plt.show()
```
